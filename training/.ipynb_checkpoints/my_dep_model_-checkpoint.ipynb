```python
import tensorflow as tf
from tensorflow.keras import models, layers
import matplotlib.pyplot as plt
import tensorflow_hub as hub
import numpy as np
from keras.layers import Dense,Flatten
```

```python
IMAGE_SIZE = 224
BATCH_SIZE = 32
CHANNELS = 3
EPOCHS = 10
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
   "Plant_Village",
    shuffle=True,
    image_size = (IMAGE_SIZE,IMAGE_SIZE),
    batch_size = BATCH_SIZE
)
```

```python
class_names = dataset.class_names
class_names
n_classes = len(class_names)
n_classes
```

```python
len(dataset)
```

```python
training_size = 0.8
size1 = int(len(dataset)*training_size)
training_dataset = dataset.take(size1)
```

```python
test_size = 0.1
size2 = int(len(dataset)*test_size)
test_dataset = dataset.skip(size1)

val_size = 0.1
size3 = int(len(dataset)*val_size)
val_dataset = test_dataset.take(size3)

test_dataset = test_dataset.skip(size3)
```

```python
print(len(training_dataset))
print(len(val_dataset))
```

```python
training_dataset = training_dataset.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
val_dataset = val_dataset.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
test_dataset = test_dataset.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
```

```python
resize_and_rescale = tf.keras.Sequential([
    layers.experimental.preprocessing.Resizing(IMAGE_SIZE,IMAGE_SIZE),
    layers.experimental.preprocessing.Rescaling(1.0/255),
]) 
```

```python
#from tensorflow.keras.applications import MobileNetV3Small
from keras.applications.mobilenet_v3 import MobileNetV3Large
```

```python
input_shape = (BATCH_SIZE,IMAGE_SIZE,IMAGE_SIZE,3)
```

```python
#base_model = MobileNetV3Small(input_shape=(IMAGE_SIZE,IMAGE_SIZE,3), weights='imagenet', include_top=False, alpha=1.0, minimalistic=True)
base_model = MobileNetV3Large(input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3), weights='imagenet', include_top=False, alpha=0.75)
#alpha se model mai kitne parameters chaiye wo change kr skte jisse fir model ke size p farak pdega..0.75 rkha hh abhi ke liye ek optimal value
```

```python
base_model.summary()
```

```python
model = models.Sequential()

model.add(resize_and_rescale)
model.add(base_model)
model.add(Flatten())
model.add(Dense(1024,activation='relu'))
model.add(Dense(1024,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(n_classes,activation='sigmoid'))

model.build(input_shape = input_shape)
```

```python
model.summary()
```

```python
for i,layer in enumerate(model.layers):
  print(i,layer.name)
```

```python
base_model.trainable = False
model.summary()
```

```python
from keras.optimizers import Adam

model.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = False),
    metrics = ['accuracy']
) 
```

```python
history = model.fit(
    training_dataset,
    epochs = EPOCHS,
    batch_size=BATCH_SIZE,
    verbose = 1,
    validation_data = val_dataset
)
```

```python
scores = model.evaluate(test_dataset)
```

```python
scores
```

```python
model_version = 2
model.save(f"../models/{model_version}")
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']
```

```python
plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(range(EPOCHS), acc, label='Training Accuracy')
plt.plot(range(EPOCHS), val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(range(EPOCHS), loss, label='Training Loss')
plt.plot(range(EPOCHS), val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
def predict(model, img):
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)
    predictions = model.predict(img_array)
    predicted_class = class_names[np.argmax(predictions[0])]
    confidence = round(100 * (np.max(predictions[0])), 2)
    return predicted_class, confidence
```
